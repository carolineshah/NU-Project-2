```python
import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import urlparse
```

```python
# import csv
path = "Resources/movie_metadata.csv"
movie_df = pd.read_csv(path)
movie_df
```

```python
# strip down to columns we want
movie_cols = ["movie_imdb_link", "movie_title", "genres", "director_name", "content_rating", "imdb_score"]
movie_df_a = movie_df[movie_cols].copy()

# rename columns
movie_df_a = movie_df_a.rename(columns={"movie_imdb_link": "IMDB_link",
                                        "movie_title": "Title",
                                        "genres": "Genres",
                                        "director_name": "Director",
                                        "content_rating": "Rating",
                                        "imdb_score": "IMDB_score"})
# drop any NaN or duplicate data
movie_df_a.dropna(how="any", inplace=True)
movie_df_a.drop_duplicates(inplace=True)

movie_df_a
```

```python
# parse imdb link for movie id
movie_id = []
for url in movie_df_a["IMDB_link"]:
    parts = urlparse(url)
    # movie_id is in the path portion so split that up
    directories = parts.path.strip('/').split('/')
    # second item in directories
    movie_id.append(directories[1])
movie_df_a["movie_id"] = movie_id
movie_df_a
```

```python
# set index to be movie_id
movie_df_a.set_index("movie_id", inplace=True)
movie_df_a
```

```python
movie_df_a.to_csv("Output/movie_data.csv")
```
